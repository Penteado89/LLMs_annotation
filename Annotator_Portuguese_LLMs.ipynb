````python
import pandas as pd
import torch
import json
import random
from transformers import AutoTokenizer, AutoConfig, LlamaForCausalLM, LlamaTokenizer, AutoModelForCausalLM, GenerationConfig
from peft import PeftModel
import textwrap

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# Definição do Template para o Prompt
FEW_SHOT_CLF_PROMPT_TEMPLATE = """
Você receberá as seguintes informações:

1. Um tweet. O tweet é delimitado com três crases.
2. Uma lista de categorias possíveis para o tweet. As categorias são: 'Favorável', 'Desfavorável', 'Neutro'.
3. Exemplos de tweets e suas categorias correspondentes.

Sua tarefa é classificar o tweet fornecido na categoria mais apropriada e fornecer um percentual de confiança na sua classificação. Não forneça nenhuma informação adicional.

Exemplos de treinamento:
```Recife vai ter uma política municipal de vacinação robusta...```, 'Favorável'
```Anvisa interrompe testes da vacina chinesa após ‘evento adverso grave’...```, 'Desfavorável'
```Os desafios científicos e tecnológicos a nossa frente...```, 'Neutro'

Exemplo para classificação: ```{x}```

Sua resposta:
"""

# Função para Construir o Prompt
def build_few_shot_prompt_slc(text: list, labels: str, template: str = FEW_SHOT_CLF_PROMPT_TEMPLATE) -> str:

    # Formatar os textos para classificar
    formatted_text = f"'''{text}'''"
    return template.format(x=text, labels=labels)

def evaluate(prompt, model, tokenizer, device):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    input_ids = inputs["input_ids"]
    # print(inputs["input_ids"])

    try:
        generation_output = model.generate(
            input_ids=input_ids,
            do_sample=True,
            temperature=0.2,
            top_p=0.75,
            num_beams=4,
            max_new_tokens=30,
            return_dict_in_generate=True,
            output_scores=True
        )

        for s in generation_output.sequences:
            output = tokenizer.decode(s)
            # Dividir a saída para obter a resposta
            parts = output.split("Sua resposta:")
            if len(parts) > 1 and parts[1].strip():  # Verificar se a resposta não está vazia
                # Retornar a categoria
                return output.split("Sua resposta:")[1].strip()
            else:
                return "Não foi possível gerar uma resposta."
    except Exception as e:
        print(f"Erro durante a geração: {e}")

    return "Não foi possível gerar uma resposta."

if __name__ == "__main__":
    # Carregar o DataFrame e preparar os dados
    # df = pd.read_csv('/content/drive/MyDrive/Anotabrita/dados_6423.csv')
    # df = df.rename(columns={'content': 'text'})
    # texto_exemplo = random.choice(df['text'].tolist())
    labels = "['Favorável', 'Desfavorável', 'Neutro']"

    # data = pd.read_csv('/content/drive/MyDrive/dados_cabrita.csv')
    # dados_treinamento_df = data.rename(columns={'content': 'text', 'positions_vac': 'label'})

    # # Gerar o prompt
    # prompt = build_few_shot_prompt_slc(texto_exemplo, labels, FEW_SHOT_CLF_PROMPT_TEMPLATE)

    # # Avaliar o prompt
    # print(prompt)
````
